```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('haberman.csv')
data.head()
```

```python
data.columns
```

```python
data['status'].value_counts()
```

```python
plt.close()
sns.set_style('whitegrid')
sns.pairplot(data, hue = 'status', size = 3)
plt.show()
```

```python
sns.FacetGrid(data, hue= 'status', size = 5)\
   .map(sns.distplot, 'age')\
   .add_legend()
plt.show()
```

```python
sns.FacetGrid(data, hue= 'status', size = 5)\
   .map(sns.distplot, 'year')\
   .add_legend()
plt.show()
```

```python
sns.FacetGrid(data, hue= 'status', size = 5)\
   .map(sns.distplot, 'number')\
   .add_legend()
plt.show()
```

```python
sns.boxplot(x = 'status', y= 'number',data = data)
plt.show()
```

```python
sns.violinplot(x = 'status', y= 'age',data = data, size = 4)
plt.show()
```

```python
data_less = data.loc[data['status'] == 1]
data_more = data.loc[data['status'] == 2]
```

```python
counts, bin_edges = np.histogram(data_less['age'], bins = 10, density = True)
pdf = counts/(sum(counts))
print(pdf)
print(bin_edges)

cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)

plt.show()
```

```python
counts, bin_edges = np.histogram(data_less['age'], bins = 10, density = True)
pdf = counts/(sum(counts))

cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)

counts, bin_edges = np.histogram(data_more['age'], bins = 10, density = True)
pdf = counts/(sum(counts))

cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf, color = 'blue')
plt.show()
```
